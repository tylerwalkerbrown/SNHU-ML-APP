### Reading back into sql 

### Query below details 
 - In this query we aggregate by year to get the:
        - Last 10 games BA
        - AVG HR vs Opp
        - Opp error 
        - AVG HRs vs opp 
        
I then split them up by the year in which the game took place and looped them over a range of years 

```python
frame = []
for year in range(2013,2023):
        query = """WITH a AS (
        SELECT
        `Opponent`,
        `Date`,
        COALESCE(SUM(H) / SUM(AB), 0) AS bat_avg,
        COALESCE(AVG(HR), 0) AS avg_HR,
        COALESCE(AVG(E), 0) AS opp_error,
        (
        SELECT COALESCE(AVG(H) / AVG(AB), 0)
        FROM college_stats.hitting_snhu h2
        WHERE h2.`Opponent` = h1.`Opponent`
        AND h2.`Date` < h1.`Date`
        AND h1.`Date` - INTERVAL 10 DAY <= h2.`Date`
        AND YEAR(h2.`Date`) = YEAR(h1.`Date`)
        ) AS 10_game_avg
        FROM
        college_stats.hitting_snhu h1
        WHERE
        YEAR(`Date`) = {year}
        GROUP BY
        `Opponent`,
        `Date`
        LIMIT
        0, 50000
        )
        SELECT DISTINCT
        a.`Date`,
        a.`Opponent`,
        a.`bat_avg`,
        a.`avg_HR`,
        a.`opp_error`,
        a.`10_game_avg`,
        college_stats.hitting_snhu.`Loc`,
        college_stats.hitting_snhu.`W/L`
        FROM
        a
        JOIN college_stats.hitting_snhu
        ON a.`Date` = college_stats.hitting_snhu.`Date`
        AND a.`Opponent` = college_stats.hitting_snhu.`Opponent`
        WHERE
        YEAR(a.`Date`) = {year}
        ORDER BY
        a.`Opponent`,
        a.`Date`
        LIMIT
        0, 50000;""".format(year = year)
        
        df = pd.read_sql(query, con=engine)
        #df = df.drop_duplicates()
        frame.append(df)
```

# Creating Links to Search for Each Teams Statistics

```python
# Getting unique values from teh hitting list for teams to concat strings in the subdomain to be search through beautiful soup
teams = list(hitting.Opponent.unique())
links = []
for team in teams:
    if team is not None:
        link = team.replace(" ", "+").replace("'", "+")
        links.append('https://www.google.com/search?q=' +link+ '+college+baseball+gamelog')
```

```python
#Searching the links on google and scraping the searches links that are found in 'div', class_='Gx5Zad fP1Qef xpd EtOod pkphOe'
import requests
from bs4 import BeautifulSoup

main_links = []

for links,teams in zip(links,teams):
    url = str(links)
    headers = {'User-Agent': 'Mozilla/5.0'}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    link_div = soup.find('div', class_='Gx5Zad fP1Qef xpd EtOod pkphOe')
    link = link_div.find('a').get('href')
    main_links.append(link) 
#    print(link)
```

```python
# Concating the subdomain to hopefully get connection to all statistics that are on teh sites page 
domain = list(pd.DataFrame(main_links)[0].str.split("/", expand = True)[3])
stats_link = []
for i in domain:     
    link = "www." + i + "/sports/baseball/stats/"
    stats_link.append(link)
```
